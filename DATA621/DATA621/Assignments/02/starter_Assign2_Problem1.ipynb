## Loading Testing dataset for the problem

```python
import pandas as pd
import numpy as np
url = "https://github.com/cfteach/NNDL_DATA621/raw/refs/heads/webpage-src/DATA621/DATA621/Assignments/02/fashion-mnist_test_rotated.csv"
test_df = pd.read_csv(url)

X_test = test_df.drop("label", axis = 1).to_numpy(dtype = np.float32)
y_test = test_df['label'].to_numpy(dtype = np.int8)

# MNIST Fashion 
labels = {0: "T-shirt/top", 1: "Trouser", 2: "Pullover", 3: "Dress", 4: "Coat",
          5: "Sandal", 6: "Shirt", 7: "Sneaker", 8: "Bag", 9: "Ankle boot"}
```

```python
import matplotlib.pyplot as plt
# plot some images
no_images_per_row = 5
no_images_per_col = 5
fig, axes = plt.subplots(no_images_per_row, no_images_per_col, figsize=(10, 10))
count = 0
for (i,j), ax in np.ndenumerate(axes):
    ax.imshow(X_test[count].reshape(28, 28), cmap="gray")
    ax.axis('off')
    ax.set_title(labels[y_test[count]])
    count+=1

plt.tight_layout()
plt.show()
```
